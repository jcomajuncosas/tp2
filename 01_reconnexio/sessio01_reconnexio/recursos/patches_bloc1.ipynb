```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio

sample_rate = 44100
```

## 1. L'eix de temps

Recordeu `range()` de TP1? `np.linspace` és l'equivalent per a nombres decimals: genera punts igualment espaiats.

```python
duration = 1.0  # segons

t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

print("Nombre de mostres:", len(t))
print("Primers 10 valors de t:", t[:10])
print("Darrers 3 valors de t:", t[-3:])
```

## 2. Generar una ona sinusoidal

`wave = np.sin(2 * pi * freq * t)` — calcula el sinus per a TOTS els valors de `t` alhora. Sense `for`.

```python
freq = 440  # Hz (un La4)

wave = np.sin(2 * np.pi * freq * t)

print("wave és un array de", len(wave), "valors")
print("Primers 10 valors:", wave[:10])
```

## 3. Visualitzar l'ona

Si dibuixéssim 44100 punts no veuríem res clar — mirem només els primers mil·lisegons.

```python
n_samples_to_plot = 500  # uns ~11ms a 44100Hz

plt.plot(t[:n_samples_to_plot], wave[:n_samples_to_plot])
plt.xlabel("temps (s)")
plt.ylabel("amplitud")
plt.title(f"Ona sinusoidal a {freq} Hz")
plt.show()
```

## 4. Escoltar el resultat

`Audio()` reprodueix l'array directament al notebook.

```python
Audio(wave, rate=sample_rate)
```

## 5. DEMO EN DIRECTE — canviem la freqüència

🎤 *Pregunta a la classe abans d'executar: si canviem `freq` a 220, sonarà més agut o més greu? I a 880?*

Prova-ho canviant el valor i tornant a executar les cel·les 2, 3 i 4 (o copia-les aquí sota).

```python
freq = 220  # prova 220, després 880

wave = np.sin(2 * np.pi * freq * t)
Audio(wave, rate=sample_rate)
```

## 6. DEMO EN DIRECTE — l'amplitud (volum)

🎤 *Pregunta: què passarà si multipliquem `wave` per 0.1? I per 2.0?*

Recordeu la Pregunta 4/5 del warm-up: `array * número` multiplica CADA element.

```python
freq = 440
amplitude = 0.1  # prova 0.1, després 1.0, després 2.0 (atenció al clipping!)

wave = amplitude * np.sin(2 * np.pi * freq * t)
Audio(wave, rate=sample_rate)
```

⚠️ Si `amplitude > 1`, el so es distorsiona ("clipping"). Per què creieu que passa? *(Pista: l'àudio s'espera entre -1 i 1)*

## 7. Empaquetar-ho en una funció

Igual que a TP1 fèieu `def transpose(note, semitones): return ...`, ara fem una funció que retorna un so sencer.

```python
def generate_tone(freq, duration, amplitude=0.5, sample_rate=44100):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    wave = amplitude * np.sin(2 * np.pi * freq * t)
    return wave

# Prova-ho:
my_tone = generate_tone(freq=330, duration=1.5, amplitude=0.3)
Audio(my_tone, rate=sample_rate)
```

## 8. Preview — altres formes d'ona

Només per escoltar la diferència de timbre. Ho treballarem a fons al Bloc 5.

```python
from scipy import signal

freq = 220
square_wave = 0.3 * signal.square(2 * np.pi * freq * t)
sawtooth_wave = 0.3 * signal.sawtooth(2 * np.pi * freq * t)
noise = 0.3 * np.random.uniform(-1, 1, len(t))

print("Quadrada:")
display(Audio(square_wave, rate=sample_rate))
print("Serra:")
display(Audio(sawtooth_wave, rate=sample_rate))
print("Soroll:")
display(Audio(noise, rate=sample_rate))
```
